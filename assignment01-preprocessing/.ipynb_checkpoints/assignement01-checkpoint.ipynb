```python
dfMaxScale = df.copy()
dfMaxScale = dfMaxScale.drop(['Heart Disease'], axis=1)
for col in dfMaxScale.columns:
        dfMaxScale[col] = dfMaxScale[col]/df[col].abs().max()
display(dfMaxScale)
```

```python
# categorical variables into quantitative variables
lebelencoder = preprocessing.LabelEncoder()
HeartDiseaseLabel = lebelencoder.fit_transform(df['Heart Disease'])
df.drop('Heart Disease', axis=1, inplace=True)
df['Heart Disease'] = HeartDiseaseLabel
df
```

```python
df['Heart Disease'] = df['Heart Disease'].map({'Presence':1, 'Absence':0}).astype(float)
df
```
